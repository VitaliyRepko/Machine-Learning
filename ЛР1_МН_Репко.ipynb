```python
a = np.array([[1, 3, -5], [0, -2, 5], [-1, 5, 2]])
b = np.array([3, 2, 2])

x = np.linalg.solve(a, b)
print("Вбудована функція: ", x)

inv_a = np.linalg.inv(a)
x1 = inv_a.dot(b)
print("Матричний метод:", x1)
```
